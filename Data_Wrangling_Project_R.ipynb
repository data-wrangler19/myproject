```r
library(dplyr) 
library(tidyverse) 
library(ggplot2)
library(visdat) 
library(skimr) 
```

```r
# For reading the csv file from its destination read_csv() function is used.
drug_df <- "C:/Users/sidhu/Desktop/Wrangling_Project/myproject-master/Accidental_Drug_Related_Deaths_2012-2018.csv" %>%
read_csv()
```

```r
# Storing the dataset in my_data_set variable.
my_data_set <- drug_df
my_data_set
```

```r
# The skim() function is used for providing a broad overview of the data a data frame.
my_data_set %>%
skim()
```

```r
#Kable() is used to get the summary statistics of dataframe with the no of observations in the dataframe.
my_data_set %>%
skim() %>%
kable()
```

```r
# glimpse () is used to see every column in data frame.
my_data_set %>%
glimpse()
```

```r
#vis_miss() is used to check if there is any missing value in the dataframe and gives the output in the graphical frame.
my_data_set %>%
vis_miss()
```

```r
# There are some columns with not applicable values in data frame which is not useful. 
# So, for dropping that columns we have used "drop_na" function.
drop_column <- my_data_set %>%
drop_na(Date,Age,DeathCity,DateType)
drop_column
```

```r
#The column sex contains some values which is "unknown", by using "drop_column" function. This function will drop the "unknown" value from the column sex.
drop_unknown_value <- with(drop_column, which(Sex == "Unknown", arr.ind = TRUE))
new_df <- drop_column[-drop_unknown_value, ]
```

```r
# For dropping the unwanted columns from the dataset the function "-c" will be use. This function will remove the unnecessary column from dataset.
drop_unwanted_column <- select(new_df,-c(ResidenceCounty,DeathCityGeo,ResidenceCityGeo,InjuryCityGeo,DeathCounty,InjuryCity,InjuryCounty,InjuryState,OtherSignifican,MannerofDeath,LocationifOther))
drop_unwanted_column
```

```r
# Some of the columns such as "sex", "race" and "residencecity" have null value which can be replace to "other" by following function.
drop_unwanted_column[c("Sex","Race","ResidenceCity")][is.na(drop_unwanted_column[c("Sex","Race","ResidenceCity")])] <- "other"
drop_unwanted_column
```

```r
# Saving the updated dataset from drop_unwanted_column to changing_name.
changing_name <-drop_unwanted_column
changing_name
```

```r
# Some values of the location column have null value which can be replaced by "unknown"
changing_name[c("Location")][is.na(changing_name[c("Location")])] <- "UnKnown"
changing_name
```

```r
# Giving the updated dataset changing_name to df_drug.
df_drug <- changing_name
df_drug
```

```r
#Replace the missing data of column DescriptionofInjury with name "Reason Unknown".
df_drug[c("DescriptionofInjury")][is.na(df_drug[c("DescriptionofInjury")])] <- "Reason UnKnown"
df_drug
```

```r
# Saving the df_drug dataset into the Final_df.
final_df <-df_drug
final_df
```

```r
# There are many blank columns in different types of drugs, we have replaced blank columns to "no".
final_df[c("Heroin","Cocaine","Fentanyl","FentanylAnalogue","Oxycodone","Oxymorphone","Ethanol","Hydrocodone","Benzodiazepine","Methadone","Amphet","Tramad","Morphine_NotHeroin","Hydromorphone","Other","OpiateNOS","AnyOpioid")][is.na(final_df[c("Heroin","Cocaine","Fentanyl","FentanylAnalogue","Oxycodone","Oxymorphone","Ethanol","Hydrocodone","Benzodiazepine","Methadone","Amphet","Tramad","Morphine_NotHeroin","Hydromorphone","Other","OpiateNOS","AnyOpioid")])] <- "No"
final_df
```

```r
# After cleaning the dataset we have got the final tidy dataset as below image with the help of vis_miss() function.
final_df %>%
vis_miss()
```

```r
# Read csv zip_code file.
zip_code <-"C:/Users/sidhu/Desktop/Wrangling_Project/myproject-master/Zipcode.csv" %>%
read_csv()
zip_code
```

```r
# We have joined the two data set which are the accident_drug and zip_code master file with the help of merge function.
join_df <- merge(x = final_df, y = zip_code, by = "ResidenceCity", all.x = TRUE)
join_df
```

```r
# using aggregate function to create count.
# And assigning the coulmn name to the values.
aggregated_df  <- aggregate(rep(1, length(join_df$Sex)),
                             by=list(join_df$Sex), sum)
```

```r
names(aggregated_df)[names(aggregated_df) == "Group.1"] <- "Gender"
names(aggregated_df)[names(aggregated_df) == "x"] <- "Death_Count"
aggregated_df
```

```r
# creating  pie chart to show the death count in percentage for different gender. 
library(ggplot2)
pie = ggplot(aggregated_df, aes(x="", y=Death_Count, fill=Gender)) + geom_bar(stat="identity")
pie = pie + coord_polar("y", start=0) + geom_text(aes(label = paste0(round((Death_Count / sum(Death_Count)) * 100, 2), "%")), position = position_stack(vjust = 0.2))
pie = pie + labs(x = NULL, y = NULL, title = "Drug overdose death count with respect to gender")
pie
```

```r
# creating new column to count the number of observations in a group.
join_df$newcol <- rep(1,nrow(join_df))
```

```r
# creating a bar plot on the basis of ResidenceState and count which is in the form of sum for death count.
join_df1 <- join_df %>%
group_by(ResidenceState)%>%
arrange(desc(ResidenceState))%>%
head(50)%>%
summarise(count = sum(newcol))%>%
ggplot(aes(x = ResidenceState, y = count)) + geom_bar(stat="identity",width=0.20, fill = "Red")+
  labs(title="Death Count with Respect to Residence State in USA", x="Residence State", y="Death Count")

join_df1
```

```r
#plotting of graph on the basis of zipcode and the death count in USA. The group_by function is used for grouping the data frame.
join_zipcode <-join_df%>%
arrange(desc(newcol))%>%
group_by(ZipCode)%>%
summarise(count = sum(newcol))%>%
head(10)%>%
ggplot(aes(x = ZipCode, y = count)) + geom_bar(stat="identity",width=0.20, fill = "Black")+
geom_point(size=4, color="orange")+
  labs(title="Death Count with Respect to Zip Code(Top 10) in descending order in USA", x="Zip Code", y="Death Count")+
 theme_bw()

join_zipcode

```

```r
# creating bar plot on the basis of Drug type and death count using ggplot.
ggplot1 <- my_desc_df %>% ggplot(aes(x = Drug_Name, y = Drug_DeathCnt)) +
  geom_bar(stat="identity", width=0.25, fill="Brown") +
  labs(title="The top 8 drug names in descending order with respect to the death count", x="Drug Name", y="Death Count")
ggplot1
```

```r
labs <- c(paste(seq(0, 99, by = 10), seq(0 + 10 - 1, 100, by = 10),
                sep = "-"))
```

```r
# convert age into different age categories and store in data frame
join_df$AgeGroup <- cut(join_df$Age, breaks = c(seq(0, 99, by = 10), Inf), labels = labs, right = TRUE)
new_df <- join_df[c("Age", "AgeGroup")]
```

```r
# counting the sum of drug accident in different age group by using aggregate function.
my_age  <- aggregate(rep(1, length(new_df$AgeGroup)), by=list(new_df$AgeGroup), sum)
my_age
```

```r
# plotting bar plot on the basis of age group and number of death count
gg_plot1 <- my_age %>% ggplot(aes(x = Group.1, y = x)) +
            geom_bar(stat = "identity",  width=0.35, fill="Light Green") +  
            labs(title="Relationship of Death Count with Respect to Age Bracket", x="Age Bracket", y="Death Count")
gg_plot1
```
