```r
# dropping na value from column date,age,deathcity and datetype
drop_column <- my_data_set %>%
drop_na(Date,Age,DeathCity,DateType)

drop_column
```

```r
#dropping the sex column with unknown entry
drop_unknown_value <- with(drop_column, which(Sex == "Unknown", arr.ind = TRUE))
NEW_df <- drop_column[-drop_unknown_value, ]
```

```r
#dropping the columns which are not needed
drop_unwanted_column <- select(NEW_df,-c(ResidenceCounty,DeathCityGeo,ResidenceCityGeo,InjuryCityGeo,DeathCounty,InjuryCity,InjuryCounty,InjuryState,OtherSignifican,MannerofDeath,LocationifOther))

drop_unwanted_column
```

```r
#Replace the missing value of sex race and city with name other
drop_unwanted_column[c("Sex","Race","ResidenceCity")][is.na(drop_unwanted_column[c("Sex","Race","ResidenceCity")])] <- "other"

drop_unwanted_column
```

```r
changing_name <-drop_unwanted_column

changing_name
```

```r
#Replace the missing value of column Location with the name Unknown
changing_name[c("Location")][is.na(changing_name[c("Location")])] <- "UnKnown"
 
changing_name
```

```r
df_drug <- changing_name

df_drug
```

```r
#Replace the missing data of column DescriptionofInjury with name reasonUnKnown
df_drug[c("DescriptionofInjury")][is.na(df_drug[c("DescriptionofInjury")])] <- "Reason UnKnown"
 
df_drug
```

```r
# read file zipcode into dataframe
zip_code <-"C:/Users/sidhu/Desktop/Wrangling_Project/myproject-master/Zipcode.csv" %>%
read_csv()

zip_code
```

```r
# joining data frame on the basis of ResidenceCity
join_df <- merge(x = Final_df, y = zip_code, by = "ResidenceCity", all.x = TRUE)


join_df
```

```r
# using aggregate function to create count for sex
aggregated_df  <- aggregate(rep(1, length(join_df$Sex)),
                             by=list(join_df$Sex), sum)
```

```r
names(aggregated_df)[names(aggregated_df) == "Group.1"] <- "Gender"
names(aggregated_df)[names(aggregated_df) == "x"] <- "Death_Count"
aggregated_df
```

```r
# creating plot to show count of death due to drug overdose in different gender 
library(ggplot2)
pie = ggplot(aggregated_df, aes(x="", y=Death_Count, fill=Gender)) + geom_bar(stat="identity")
pie = pie + coord_polar("y", start=0) + geom_text(aes(label = paste0(round((Death_Count / sum(Death_Count)) * 100, 2), "%")), position = position_stack(vjust = 0.2))
pie = pie + labs(x = NULL, y = NULL, title = "Drug overdose death count with respect to gender")
pie
```

```r
# creating new column to count the number of observations in a group
join_df$newcol <- rep(1,nrow(join_df))
```

```r
# creating a bar plot on the basis of ResidenceState and count the sum of accident due to drug
join_df1 <- join_df %>%
group_by(ResidenceState)%>%
arrange(desc(ResidenceState))%>%
head(50)%>%
summarise(count = sum(newcol))%>%
ggplot(aes(x = ResidenceState, y = count)) + geom_bar(stat="identity",width=0.20, fill = "Red")+
  labs(title="Death Count with Respect to Residence State in USA", x="Residence State", y="Death Count")

join_df1
```

```r
#plotting of graph on the basis of zipcode and the death count in USA
join_ZipCode <-join_df%>%
arrange(desc(newcol))%>%
group_by(ZipCode)%>%
summarise(count = sum(newcol))%>%
head(10)%>%
ggplot(aes(x = ZipCode, y = count)) + geom_bar(stat="identity",width=0.20, fill = "Black")+
geom_point(size=4, color="orange")+
  labs(title="Death Count with Respect to Zip Code(Top 10) in descending order in USA", x="Zip Code", y="Death Count")+
 theme_bw()

join_ZipCode

```

```r
labs <- c(paste(seq(0, 99, by = 10), seq(0 + 10 - 1, 100, by = 10),
                sep = "-"))
```

```r
# convert age into different age categories and store in data frame
join_df$AgeGroup <- cut(join_df$Age, breaks = c(seq(0, 99, by = 10), Inf), labels = labs, right = TRUE)
new_df <- join_df[c("Age", "AgeGroup")]
```

```r
# counting the sum of drug accident in different age group
my_age  <- aggregate(rep(1, length(new_df$AgeGroup)), by=list(new_df$AgeGroup), sum)
my_age
```

```r
# plotting bar plot =on the basis of age group and number of death count
gg_plot1 <- my_age %>% ggplot(aes(x = Group.1, y = x)) +
            geom_bar(stat = "identity",  width=0.35, fill="Light Green") +  
            labs(title="Relationship of Death Count with Respect to Age Bracket", x="Age Bracket", y="Death Count")
gg_plot1
```
