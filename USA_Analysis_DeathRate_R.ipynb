```r
# Loading the required package for the project
```

```r
library(tidyverse)
```

```r
# Loading the csv file into a dataframe object
```

```r
my_df <- read_csv("G:/SEMESTER TWO/Data Wrangling/Project/myproject-master/myproject-master/death-rates-from-drug-use-disorders.csv")
my_df
```

```r
# Filtering the latest year in the file which is 2017 and showing the first 10 records
```

```r
latest_df <- my_df %>% filter(Year == "2017")
latest_df %>% head(10)
```

```r
# Grouping the dataframe with respect to the Entity column and showing the top 20 records
```

```r
grouped_df <- latest_df %>% group_by(Entity)
grouped_df %>% head(20)
```

```r
# Renaming the column name for Death Rate
```

```r
names(grouped_df)[names(grouped_df) == "Deaths - Drug use disorders - Sex: Both - Age: Age-standardized (Rate) (deaths per 100,000 individuals)"] <- "Death_Rate"
grouped_df
```

```r
# Removing the unwanted records as per our analysis with respect to the Entity name
```

```r
drop_unknown_value <- with(grouped_df, which(Entity == "High-income", arr.ind = TRUE))
dropped_df <- grouped_df[-drop_unknown_value, ]
drop_unknown_value1 <- with(dropped_df, which(Entity == "High SDI", arr.ind = TRUE))
dropped_df_final <- dropped_df[-drop_unknown_value1, ]
dropped_df_final
```

```r
# Arranging the above dataframe in descending order of death rate and showing the first six records
```

```r
my_countwise_df <- dropped_df_final %>% arrange(desc(Death_Rate)) %>% head(6)
my_countwise_df
```

```r
# Plotting the above dataframe table in bar graph showing the top six countries with maximum death rate for the year 2017
```

```r
gg_plot1 <- my_countwise_df %>% ggplot(aes(x = Entity, y = Death_Rate)) +
            geom_bar(stat = "identity",  width=0.20, fill="Sky Blue") +  
            labs(title="Relationship of Death Count in Descending order for Top 6 Countries(Year-2017)", x="Country Name", y="Death Rate Per 100,000 Individuals")
gg_plot1
```
